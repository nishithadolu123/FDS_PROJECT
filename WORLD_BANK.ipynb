```python
from sklearn.linear_model import LinearRegression
```

```python
# Clean column names
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
# Reshape data (years become rows)
df_long = df.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Population')
df_long.dropna(inplace=True)
df_long['Year'] = df_long['Year'].astype(int)

# Filter for a specific country (e.g., India)
country = "India"  # Change to "United States", "China", etc.
country_df = df_long[df_long['Country Name'] == country]
```

## üßπ Data Cleaning

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
```

## üìà Extracting Population of 2021

```python
#Plotting the top 10 countries by population in 2021
df_2021 = df[['Country Name', '2021']]
df_2021 = df_2021.dropna()
df_2021.columns = ['Country', 'Population']
```

```python
print("‚úÖ Extracted 2021 Population Data (first 10 rows):")
display(df_2021.head(10))  # Display first 10 rows of cleaned data
```

## üåü Top 10 Countries by Population in 2021

## üìä Data Visualizations :

```python
# Line plot of Population over time
plt.figure(figsize=(10, 5))
sns.lineplot(data=country_df, x='Year', y='Population', marker='o')
plt.title(f"{country} - Population Growth (1960‚Äì2023)")
plt.ylabel("Population")
plt.grid(True)
plt.show()
```

```python
# Predict Population for 2024 & 2025
model = LinearRegression()
X = country_df[['Year']]
y = country_df[['Population']]
model.fit(X, y)
future_years = pd.DataFrame({'Year': [2024, 2025]})
predictions = model.predict(future_years)

print(f"\nüìà Predicted Population for {country}:")
for year, pred in zip(future_years['Year'], predictions):
    print(f"{year}: {int(pred[0]):,}")
```

```python
# Top 10 Countries by Population in 2023
latest_year = 2023
top_df = df_long[df_long['Year'] == latest_year].nlargest(10, 'Population')

plt.figure(figsize=(10, 6))
sns.barplot(x='Population', y='Country Name', data=top_df, palette="viridis")
plt.title(f"Top 10 Countries by Population in {latest_year}")
plt.show()
```

```python
#  Heatmap of Population Growth (Top 10 Countries)
top_countries = df_long[df_long['Year'] == 2023].nlargest(10, 'Population')['Country Name'].tolist()
heatmap_df = df[df['Country Name'].isin(top_countries)].set_index('Country Name')
heatmap_df = heatmap_df.drop(columns=['Country Code'])
heatmap_df = heatmap_df.loc[:, '2000':'2023']

plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_df.astype(float), cmap='YlGnBu', linewidths=0.5)
plt.title("Population Heatmap (Top 10 Countries, 2000‚Äì2023)")
plt.xlabel("Year")
plt.ylabel("Country")
plt.show()
```

```python
# Box Plot of Population for Top 5 Countries (2000‚Äì2023)
box_df = df_long[df_long['Country Name'].isin(top_countries) & (df_long['Year'] >= 2000)]

plt.figure(figsize=(10, 6))
sns.boxplot(x='Country Name', y='Population', data=box_df)
plt.xticks(rotation=45)
plt.title("Population Box Plot (2000‚Äì2023)")
plt.show()
```

```python
# ‚úÖ HISTOGRAM: Distribution of Population in 2023
plt.figure(figsize=(10, 6))
sns.histplot(df['2023'], bins=30, kde=True, color='coral')
plt.title("Distribution of Population Across Countries (2023)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.grid(True)
plt.show()
```

```python
# ‚úÖ PAIR PLOT: Compare Population across selected years
selected_years = ['2000', '2010', '2020', '2023']
pair_df = df[selected_years].dropna()

sns.pairplot(pair_df)
plt.suptitle("Pair Plot: Population Comparisons Across Years", y=1.02)
plt.show()
```

```python
# ‚úÖ PIE CHART: Top 5 Countries by Population in 2023
top5_2023 = df[['Country Name', '2023']].dropna().nlargest(5, '2023')
others = df[['Country Name', '2023']].dropna().nsmallest(len(df) - 5, '2023')['2023'].sum()

# Combine top 5 + others
labels = top5_2023['Country Name'].tolist() + ['Others']
sizes = top5_2023['2023'].tolist() + [others]

# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title("World Population Share (Top 5 Countries in 2023)")
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.show()
```

## Prediction

```python
# Select a country (change this as needed)
country_name = "India"

# Extract the row for that country
country_row = df[df["Country Name"] == country_name]
```

```python
# Prepare data: Years (X) and Population (y)
years = np.array([int(col) for col in df.columns if col.isdigit()]).reshape(-1, 1)
population = country_row.loc[:, country_row.columns.str.isnumeric()].values.flatten()
```

```python
# Drop any missing values if present
valid_indices = ~np.isnan(population)
X = years[valid_indices]
y = population[valid_indices].reshape(-1, 1)
```

```python
# Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict population for 2024 and 2025
future_years = np.array([[2024], [2025]])
future_predictions = model.predict(future_years)
```

```python
# Print predictions
print(f"\nüìà Population Predictions for {country_name}:")
for year, pop in zip(future_years.flatten(), future_predictions.flatten()):
    print(f"{year}: {int(pop):,}")
```

```python
#  Plot actual + predicted
plt.figure(figsize=(10, 5))
plt.plot(X, y, label='Actual Population', marker='o')
plt.plot(future_years, future_predictions, label='Predicted Population (2024‚Äì2025)', marker='x', linestyle='--', color='red')
plt.title(f"Population Prediction for {country_name}")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid(True)
plt.show()
```
